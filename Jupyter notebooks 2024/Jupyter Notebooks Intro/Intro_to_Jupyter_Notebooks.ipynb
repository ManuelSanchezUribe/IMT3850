Primero, definiremos una función que simule el dado cargado:

````python
def dado_cargado(x):
    '''
    Para x en [0,1], retorna un valor del dado.
    Si se entrega otro valor, entonces no es un dado (le asignamos 0 solo por que sí)
    '''
    if 0 <= x < 0.1:
        return 1
    elif 0.1 <= x < 0.3:
        return 2
    elif 0.3 <= x <0.6:
        return 3
    elif 0.6 <= x < 0.8:
        return 4
    elif 0.8 <= x < 0.9:
        return 5
    elif 0.9 <= x <= 1:
        return 6
    else:
        return 0
````

Hacemos un sampling uniforme sobre [0,1] y le asignamos del dado correspondiente a cada uno:

````python
x = np.random.random(10000)
y = np.array([dado_cargado(i) for i in x])

print(x[:10])
print(y[:10])
````

### Ejercicio 5
La iteración de punto fijo es un método para encontrar solución a funciones de manera numérica dadas ciertas condiciones de regularidad en la función. Este consiste en reescribir la ecuación $f(x) = 0$ en una ecuación de la forma $x = g(x)$. En este problema usaremos la función

$$f(x) = x^2-x$$

Entonces, se hará lo siguiente:

$$f(x) = 0 \Leftrightarrow -x + x^2 = 0$$
$$\Leftrightarrow x = x^2$$

Entonces, tenemos $g(x) = x^2$. Luego, la siguiente iteración será aquella que nos llevará a las raíces de las soluciones:

$$x_{k+1} = g(x_k)$$

Pero como elejimos un $x_0$ tal que el método converja? Para esto se tienen los siguientes pasos:

1. Ver gráfica función original $f$
2. Elegir un punto cercano a la raíz como $x_0$

<b>Problema:</b> Grafique el vector $\{x_k\}_{k=0}^{10}$ para la raíz $0$ con 3 valores iniciales distintos y compare la rapidez del método. Verifique también que si $x_0 > 1$ entonces el método no converge. Por que? A continuación se muestra la gráfica de la función $f$:


````python
f = lambda x: x**2-x
x = np.linspace(-1,2, num=100)


plt.figure(figsize=(8,6))
plt.title("Gráfica de f(x)", fontsize=16)
plt.plot(x, f(x), label='$f$(x)', ls="--")
plt.vlines(x=0, ymin=-1/4, ymax=2, color="black", lw=2, alpha=0.3)
plt.hlines(y=0, xmin=-1, xmax=2, color="black", lw=2, alpha=0.3)
plt.scatter([0, 1], [0, 0], color="red", marker="*", s=100, label="raíces de $f$")
plt.legend(loc='best', fontsize=12)
plt.show()
````

### Solución:

````python
g = lambda x: x**2
x1 = np.zeros(11)
x2 = np.zeros(11)
x3 = np.zeros(11)
x1[0] = 0.1
x2[0] = 0.5
x3[0] = 0.99
for k in range(1,11):
    x1[k] = g(x1[k-1])
    x2[k] = g(x2[k-1])
    x3[k] = g(x3[k-1])
````

````python
y = np.arange(11) # = np.array(range(11))
plt.plot(y, x1, 'b*')
plt.plot(y, x2, 'r:')
plt.plot(y, x3, 'g--')
plt.show()
````

````python
x4 = np.zeros(11)
x4[0] = 1.1
for k in range(1,11):
    x4[k] = g(x4[k-1])

plt.plot(y, x4, 'bo')
plt.yscale('log')
plt.show()
````

Veremos algunos ejemplos de estos:

````python
from sklearn import linear_model
import seaborn as sns

sns.set(rc={'figure.figsize':(8, 6)})

data_1 = pd.read_csv("data1.csv", index_col=0)
````

Aquí, ``data_1`` almacena puntos en $\mathbb{R}^2$ y su etiqueta o *label*.

````python
data_1
````

Veamos como son estos:

````python
ax = sns.scatterplot(data=data_1, x="x", y="y", style="label", hue='label')
````

````python
X = data_1[["x", "y"]].values
y = data_1[["label"]].values.ravel()
clasificador = linear_model.Perceptron(fit_intercept=True)
````

Entrenamos el perceptron:

````python
clasificador.fit(X, y)
````

````python
# Aquí vemos los coeficientes e intercepto de nuestro clasificador
# Estos son  tal que se cumple la ecuación: Ax + By + C = 0

# Coeficientes de los atributos 
for atr in range(clasificador.n_features_in_):
    print("El atributo ", atr, " tiene asociado el coeficiente ", clasificador.coef_[0, atr])

# Intercepto
print(f"El intercepto es: {clasificador.intercept_[0]}")
````

Graficamos el resultado de nuestro clasificador entrenado con los datos:

````python
X_plot = np.linspace(-3, 3, 1000)
y_plot = -1*(clasificador.coef_[0, 0]*X_plot + clasificador.intercept_[0])/clasificador.coef_[0, 1]

## AX+BY+C --> Y = (-AX - C)/B 

ax = sns.scatterplot(data=data_1, x="x", y="y", style="label", hue='label')
plt.plot(X_plot, y_plot, color='r', label="Clasificador") #Grafico de la linea de perceptron
plt.legend(loc="best")
plt.show()
````

````python
Xpred = [[0, -5], [3, -10], [-3, 8], [0, 0]] # Debiese ser [0, 1, 0, 1]

print(clasificador.predict(Xpred))
````
